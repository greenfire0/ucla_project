```python
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python

 
target=pd.read_csv("guns.csv")

target_time_used=target.groupby(by="weapon")["time_used"].sum().sort_values(ascending=False)
target_weapon_kills=target.groupby(by="weapon")["knocks"].mean().sort_values(ascending=False)

print(target_weapon_kills)
target_weapon=target_time_used.index
target_weapon2=target_weapon_kills.index
print(target_weapon2)
```

```python
textprops = {"fontsize":10} #label's fontsize
plt.title("Time used by different guns", bbox={'facecolor':'0.8', 'pad':5})
plt.pie(target_time_used,labels=target_weapon,radius=3.5,shadow=True,autopct='%1.1f%%',counterclock=False,startangle=90,textprops=textprops)#change the radius if want a different size of pie chart
```

```python
font1={'fontsize': 20,'verticalalignment': 'baseline'}
plt.title("Mean kills by different guns", bbox={'facecolor':'0.8', 'pad':5})
plt.pie(target_weapon_kills,labels=target_weapon2,radius=3.5,shadow=True,counterclock=False,startangle=90,autopct='%1.1f%%',textprops=textprops)#change the radius if want a different size of pie chart
```
