```python
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
#read json file
with open('test_2.json') as json_data:#change the json name
    data = json.load(json_data)

target = pd.DataFrame(data['games'])

'''
#use this if it is a csv file 
target=pd.read_csv("out.csv")
'''
target_num_people=target["kills"].value_counts()
print(target_num_people)
target_num_kills=target_num_people.index
print(target_num_kills)
```

```python
textprops = {"fontsize":15} #label's fontsize
plt.pie(target_num_people,labels=target_num_kills,radius=3,shadow=True,autopct='%1.1f%%',textprops=textprops)#change the radius if want a different size of pie chart
```
